```python
# Primeiro vou separar os anos que estão no padrão YYYY ~ YYYY em duas colunas diferentes, chamadas Contract Start e Contract End
# Depois criar uma nova coluna chamada Contract Lenght (Years), calculando a duração do contrato
# Depois atribuir um NaN aos contratos que não tem informação de datas e criar uma última coluna com o Tipo do contrato
```

```python
def extract_contract_info(contract):
  if contract == 'Free' or 'On Loan' in contract:
    start_year = np.nan
    end_year = np.nan
    contract_length = 0
  else:
    start_date, end_date = contract.split(' ~ ')
    start_year = int(float((start_date[0:4])))
    end_year = int(float((end_date[0:4])))
    contract_length = end_year - start_year
  return start_year, end_year, contract_length
```

```python
# Aplicar a função na coluna de contratos para fazer a limpeza dos dados
novas_colunas = ['Contract Start', 'Contract End', 'Contract Length (years)']
novos_dados = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))
novos_dados
```

```python
# Iterar para adicionar os novos dados nas novas colunas criadas
for i in range (len(novas_colunas)):
  fifa.insert(loc=fifa.columns.get_loc('Contract') + 1 + i, column=novas_colunas[i], value=novos_dados[i].round(0))
```

```python
# Por algum motivo que eu ainda não entendi, os anos estão em float, hora de colocar em int
fifa['Contract Start'] = fifa['Contract Start'].astype('float').astype('Int64')
fifa['Contract End'] = fifa['Contract End'].astype('float').astype('Int64')
fifa['Contract Length (years)'] = fifa['Contract Length (years)'].astype('float').astype('Int64')
pd.set_option('display.max_columns', None)
fifa.sample(5)
```

```python
# Criar uma função para categorizar os contratos
def categorizar_contratos(contract):
  if contract == 'Free':
    return 'Free'
  elif 'On Loan' in contract:
    return 'On Loan'
  else:
    return 'Active Contract'

```

```python
# Criar a coluna de Contract Status e adicionar os dados

fifa.insert(fifa.columns.get_loc('Contract')+1, column='Contract Status', value=fifa['Contract'].apply(lambda x: categorizar_contratos(x)))
```

```python
# As alturas dos jogadores estão em valores diferentes (polegadas, centímetros e em strings. Devemos mudar todos para cm e transformar em int)
fifa['Height'].unique()
```

```python
def conversor_altura(altura):
  if 'cm' in altura:
    return int(altura.strip('cm'))
  else:
    # Separar os pés e polegas e depois somar para transformar em centímetros
    feet, inches = altura.split("'")
    total_inches = int(feet)*12 + int(inches.strip('"'))
    return round(total_inches * 2.54)
```

```python
# Aplicando a função para alterar todas as alturas
fifa['Height'] = fifa['Height'].apply(lambda x: conversor_altura(x))
```

```python
#Confirmando que deu tudo certo
fifa['Height'].unique()
```

```python
#Alterando o nome da coluna para dar clareza a unidade de medida utilizada
fifa = fifa.rename(columns={'Height': 'Height (cm)'})
fifa.sample(3)
```

```python
# Definindo uma função de conversão de peso

def conversor_peso(peso):
  if 'kg' in peso:
    return int(peso.strip('kg'))
  else:
    return round(int(peso.strip('lbs')) * 0.453592)
```

```python
fifa['Weight'] = fifa['Weight'].apply(lambda x: conversor_peso(x))
fifa['Weight'].unique()
```

```python
# O Padrão da coluna valores será M de Euros. Criando função para limpar os dados


def conversor_valores(valor):
  valor = valor.strip('€')
  if 'K' in valor:
    valor = valor.strip('K')
    valor = float(valor) / 100
  elif 'M' in valor:
    valor = float(valor.strip('M'))
  return float(valor)
```

```python
fifa['Value'] = fifa['Value'].apply(lambda x: conversor_valores(x))
```

```python
# Como são quantidades parecidos, vou fazer em milhares
def conversor_multas(multa):
  multa = multa.strip('€')
  if 'K' in multa:
    multa = multa.strip('K')
  if 'M' in multa:
    multa = multa.strip('M')
    multa = float(multa)*100
  return float(multa)
```

```python
# Hits é a quantidade de pesquisas que fizeram por um jogador. Esse número está como object pois em alguns casos tem um "K" em referência
# a milhares de pesquisa. Definir uma função para transformá-los em int no final.

def conversor_hits(hits):
  if 'K' in hits:
    hits = hits.strip('K')
    hits = float(hits)
    hits = hits * 1000
  else:
    hits = int(hits)
  return int(hits)
```

```python
fifa['Hits'] = fifa['Hits'].apply(lambda x: conversor_hits(x))
```
