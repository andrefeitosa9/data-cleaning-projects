Legenda para a planilha:

Survived (Sobreviveu): 0 = Não, 1 = Sim

Pclass (Classe): Classe de ingresso 1 = 1º, 2 = 2º, 3 = 3º

Sex (Sexo): Sexo do passageiro

Age (Idade): Idade em anos

Sibsp: Quantidade de irmãos / cônjuges a bordo do Titanic

Parch: Quantidade de pais / crianças a bordo do Titanic

Ticket (Bilhete): Número do bilhete de embarque

Fare (Tarifa): Tarifa paga pelo Passageiro

Cabin (Cabine): Número de cabine

Embarked (Embarque): Porto de Embarque (C = Cherbourg, Q=Queenstown, S = Southampton)

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(r'/content/train (1).csv')
data
```

```python
# Primeiras impressões:
# Dividir a coluna nome em "Título", 'First Name', 'Last Name'
# Mudar Age para int
# Tratar os NaN em "Cabin"
```

```python
data.info()
# Existem 177 missing values em Age.
# Existem 2 missing values em Embarked
```

```python
data.duplicated().sum()
# Já que não há duplicados, seguimos.
```

```python
def separador_nome(nome):
  last_name, title_and_first_name = nome.split(',')
  title, first_name = title_and_first_name.split(". ", 1)
  return title, first_name, last_name
```

```python
# Aplicar a função na coluna de contratos para fazer a limpeza dos dados
nomes_separados = data['Name'].apply(lambda x: pd.Series(separador_nome(x)))
nomes_separados
```

```python
novas_colunas = ['Title', 'First Name', 'Last Name']

# Iterar para adicionar os novos dados nas novas colunas criadas

for i in range (len(novas_colunas)):
  data.insert(loc=data.columns.get_loc('Name') + 1 + i, column = novas_colunas[i], value=nomes_separados[i])

data.head(10)
```

```python
# Após checagem no Google, ambos passageiros embarcaram em Southampton. Atualizando os dados:

data.loc[data.Embarked.isnull(), 'Embarked'] = 'S'
data.info()
```

```python
# Criar uma mediana de idade baseada em Pclass e Gênero para aplicar nos missing values de Age

data.groupby(['Pclass', 'Sex'])['Age'].median()

data['Age'] = data.Age.fillna(data.groupby(['Pclass', 'Sex'])['Age'].transform('median'))
```

```python
# Atribuindo a sigla 'NR' (No Register) nos valores nulos de Cabin

data['Cabin'] = data['Cabin'].fillna('NR')
```

```python
data.info()
data.to_csv('Titanic Data Cleaned.csv')
```
